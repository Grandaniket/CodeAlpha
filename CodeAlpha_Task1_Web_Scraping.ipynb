**Objective** -

The objective of this project is to use python library like BeautifulSoup to extract data from a public website, identify and collect relevant data from web pages, understand the HTML structure and web navigation to gather accurate data and create a custom dataset for further analysis.

**Understanding the Website** -

The website used in this project is https://quotes.toscrape.com/. It's a practice website designed for learning web scraping techniques. It contains famous quotes from different authors. It includes three pieces of information that is, the quote itself, author's name and tags.  
The quotes are distributed across multiple pages, with 10 quotes on each page. This structure makes the website suitable for practicing page navigation and data extraction.

```python
# Importing libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Loading and checking the website

website_url = "https://quotes.toscrape.com/"

response = requests.get(website_url)
print(response.status_code)
```

**Observation** -

Status code 200 means the website was downloaded successfully.

```python
# Parsing HTML

soup = BeautifulSoup(response.text,"html.parser")
```

**Observation** -

`BeautifulSoup` module converts the HTML into a format that Python can easily search.

```python
# Extracting container

quotes = soup.find_all("div", class_= "quote")

print(len(quotes))
```

**Observation** -

The `find_all()` function searches the HTML page and finds all `<div>` elements with the class name `quote` , whereas, the `len()` function counts the total number of quote containers found on the current page.    
The first page contains 10 quotes. Each quote is stored inside.

```python
# Creating empty list

all_quotes = []
```

**Observation** -

An empty list named `all_quotes` is created to store the scraped data.

```python
# Scraping data 

page_number = 1
while True:
    
    website_url = f"https://quotes.toscrape.com/page/{page_number}/"
    response = requests.get(website_url)
    
    if response.status_code != 200 :
        break

    soup = BeautifulSoup(response.text, "html.parser")
    quotes = soup.find_all("div", class_= "quote")

    if len(quotes) == 0 :
        break

    for quote in quotes :

        quotations = quote.find("span", class_= "text").text
        authors = quote.find("small", class_= "author").text
        tags = ", ".join([t.text for t in quote.find_all("a", class_= "tag")])

        all_quotes.append([quotations, authors, tags])

    page_number += 1
```

**Observation** -

The scraper starts from page 1 and downloads each webpage using the `requests` library. BeautifulSoup parses the HTML and finds all quote sections using the `<div class = "quote">` tag.  
For each quote on the webpage, the program extracts three pieces of information :  
1. Quote Text : The actual quote, stored inside the `<span class="text">` element.  
2. Author Name : The name of the author, stored inside the `<small class="author">` element.  
3. Tags : The keywords related to the quote, stored inside one or more `<a class="tag">` elements.
   
The extracted quote, author, and tags are then added to the `all_quotes` list. After all pages have been scraped, this list contains the complete dataset and is ready to be converted into a pandas DataFrame.  
After completing one page, the scraper automatically moves to the next page. The process continues until there are no more pages to scrape. At that point, the loop stops automatically.

```python
print(len(all_quotes))
```

**Observation** -

The website contains 100 quotes in total. All quotes have now been collected.

```python
# Creating pandas dataframe

df = pd.DataFrame(all_quotes, columns = ["Quote", "Author", "Tags"])
df.head()
```

**Observation** -

A dataframe has been created. We can see the first five rows.

```python
df.tail()
```

**Observation** -

The last five rows of the dataset have been displayed and we can see that the data has been scraped successfully until the last page of the website.

```python
# Saving the dataset

df.to_csv("Quotes.csv", index = False)

print("CSV file saved successfully.")
```
